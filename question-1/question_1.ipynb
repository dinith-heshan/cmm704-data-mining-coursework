Setup environment variables and HTTP authentification token

```python
import json

with open('/root/.kaggle/kaggle.json') as f:
    kaggle = json.load(f)
```

```python
from base64 import b64encode

USERNAME = kaggle['username']
KEY = kaggle['key']

AUTH = b64encode(f"{USERNAME}:{KEY}".encode()).decode()
AUTH_HEADER = f"Authorization: Basic {AUTH}"
```

```python
import os

os.environ['USERNAME'] = USERNAME
os.environ['KEY'] = KEY
os.environ['AUTH'] = AUTH
os.environ['AUTH_HEADER'] = AUTH_HEADER

os.environ['DATASET_URL'] = "https://www.kaggle.com/api/v1/competitions/data/download/diabetic-retinopathy-detection"
```
